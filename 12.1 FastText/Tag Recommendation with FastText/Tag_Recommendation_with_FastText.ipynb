```python
import pandas as pd
df = pd.read_csv("datasets/Myntra Products/products.csv")
```

```python
# Features
df['features'] = df['product_name'] + ' ' + df['product_tag'] + ' '+ df['brand_tag']
# Labels
# Adding comma at front to split all and join it with label
df['tags'] = ','+df['tags']
df['label'] = ['__label__'.join(tag.split(',')).strip().lower() for tag in df['tags']]

df.head(3)
```

```python
data = [i[-1]+ ' ' + i[-2] for i in df.values]
fd = open('data.txt','w')
for i in data:
  fd.write(i)
  fd.write('\n')
fd.close()
```

```python
import fasttext
model = fasttext.train_supervised(input='data.txt')
```

```python
import numpy as np
for i in range(10):
  # Take random feature every time
  txt = df['features'][np.random.randint(len(df['features']))]
  print(txt)
  # Return 5 closest recommendations
  print(model.predict(txt.lower(),k=5))
```
